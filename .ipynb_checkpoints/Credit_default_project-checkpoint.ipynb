Очистка и подготовка данных

```python
df_train = pd.read_csv('train.csv', sep = ',', encoding = 'utf8')
df_test = pd.read_csv('test.csv', sep = ',', encoding = 'utf8')

#заполняем пропуски Years in current job на '10+ years' (типа это медиана)
df_train['Years in current job'] = df_train['Years in current job'].fillna('10+ years')
df_test['Years in current job'] = df_test['Years in current job'].fillna('10+ years')

#заменяем текстовые значения Years in current job на числовые
df_train['Years in current job'].replace({'10+ years': 10, 
        '9 years': 9, '8 years' : 8, '7 years' : 7, '6 years' : 6, '5 years' : 5, 
        '4 years' : 4, '3 years' : 3, '2 years' : 2, '1 year' : 1, '< 1 year' : 0}, inplace=True)
df_test['Years in current job'].replace({'10+ years': 10, 
        '9 years': 9, '8 years' : 8, '7 years' : 7, '6 years' : 6, '5 years' : 5, 
        '4 years' : 4, '3 years' : 3, '2 years' : 2, '1 year' : 1, '< 1 year' : 0}, inplace=True)

#удаляем слишком высокие значения Maximum Open Credit
df_train = df_train.drop(df_train[df_train['Maximum Open Credit'] >= 100000000].index)

#заменяем Current Loan Amount = 99999999 на медиану 
df_train['Current Loan Amount'] = np.where(df_train['Current Loan Amount'] == 99999999, 
                             np.median(df_train['Current Loan Amount']), df_train['Current Loan Amount'])
df_test['Current Loan Amount'] = np.where(df_test['Current Loan Amount'] == 99999999, 
                             np.median(df_train['Current Loan Amount']), df_test['Current Loan Amount'])

#заполняем пропуски Annual Income медианой
df_train['Annual Income'] = df_train['Annual Income'].fillna(np.nanmedian(df_train['Annual Income']))
df_test['Annual Income'] = df_test['Annual Income'].fillna(np.nanmedian(df_train['Annual Income']))

#заполняем пропуски Months since last delinquent медианой
df_train['Months since last delinquent'] = df_train['Months since last delinquent'].\
            fillna(np.nanmedian(df_train['Months since last delinquent']))
df_test['Months since last delinquent'] = df_test['Months since last delinquent'].\
            fillna(np.nanmedian(df_train['Months since last delinquent']))

#заполняем пропуски Bankruptcies медианой
df_train['Bankruptcies'] = df_train['Bankruptcies'].fillna(np.nanmedian(df_train['Bankruptcies']))
df_test['Bankruptcies'] = df_test['Bankruptcies'].fillna(np.nanmedian(df_train['Bankruptcies']))

#заполняем пропуски Credit Score медианой
df_train['Credit Score'] = df_train['Credit Score'].fillna(np.nanmedian(df_train['Credit Score']))
df_test['Credit Score'] = df_test['Credit Score'].fillna(np.nanmedian(df_train['Credit Score']))

#новая переменная - доля месячного долга от годового дохода
df_train['Burden'] = df_train['Monthly Debt']/(df_train['Annual Income']/12)
df_test['Burden'] = df_test['Monthly Debt']/(df_test['Annual Income']/12)

#новая переменная - использована ли кредитная линия или еще нет
df_train['Used index'] = np.where(df_train['Current Credit Balance'] == 0, 1, 0)
df_test['Used index'] = np.where(df_test['Current Credit Balance'] == 0, 1, 0)

#новая переменная - доля использованного кредита в доходе
df_train['Used credit to income'] = df_train['Current Loan Amount']/df_train['Annual Income']
df_test['Used credit to income'] = df_test['Current Loan Amount']/df_test['Annual Income']

#новая переменная - доля всего кредита в доходе
df_train['Total credit to income'] = (df_train['Current Loan Amount'] + df_train['Current Credit Balance'])/ \
                                df_train['Annual Income']
df_test['Total credit to income'] = (df_test['Current Loan Amount'] + df_test['Current Credit Balance'])/ \
                                df_test['Annual Income']

#новая переменная - ставка процента
df_train['Proxy interest rate'] = df_train['Monthly Debt']/df_train['Current Loan Amount']*12
df_test['Proxy interest rate'] = df_test['Monthly Debt']/df_test['Current Loan Amount']*12



#кодируем строковые переменные
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

names = ['Home Ownership', 'Purpose', 'Term']

def encoding(data_train, data_test, names):
    le_names = [i +' le' for i in names]
    for i in range(len(names)):
        le.fit(data_train[names[i]])
        data_train[le_names[i]] = le.transform(data_train[names[i]])
        data_test[le_names[i]] = le.transform(data_test[names[i]])

encoding(df_train, df_test, names)

#df_train[['Home Ownership', 'Years in current job', 'Purpose', 'Term']].unique()
df_train.head(5)
```

```python
features = df_test.columns.drop(names)

def compare(feature):

    plt.hist(df_train[feature], bins=50, alpha  = 0.5, color = 'green', label='train')
    plt.hist(df_test[feature], bins=50, alpha  = 0.5, color = 'red', label='test')
    plt.title(feature)
    plt.legend()
    plt.yscale('log')
    plt.show()

for feature in features:    
    compare(feature)
```

Подготовка модели

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

target = 'Credit Default'
# 'Years in current job', 'Tax Liens', 'Number of Credit Problems', 'Bankruptcies', 'Used index',
features = ['Annual Income',  'Number of Open Accounts', 'Years of Credit History', 
       'Maximum Open Credit', 'Months since last delinquent', 
       'Current Loan Amount', 'Current Credit Balance', 'Monthly Debt', 'Credit Score', 
       'Home Ownership le', 'Purpose le', 'Term le', 
       'Burden', 'Used credit to income', 'Total credit to income', 'Proxy interest rate']

y = df_train[target].values.ravel()
x = df_train[features].values

param_grid  = {'base_estimator__max_depth':[2, 5, 7, 10],
               'base_estimator__min_samples_split':[2],
               'base_estimator__min_samples_leaf':[1],
               'base_estimator__n_estimators':[100, 300]}

dt = DecisionTreeClassifier(criterion = 'gini', # 'entropy'
                             splitter = 'best', # 'random' 
                            class_weight = 'balanced'
                             random_state = 1)

ada = AdaBoostClassifier(base_estimator = dt,
                         learning_rate = 1,
                         #algorithm='SAMME.R',
                         random_state = 2)

grid = GridSearchCV(ada, param_grid = param_grid, cv = 10, 
                    scoring = 'f1') 

grid.fit(x, y)

print(grid.best_estimator_)

print(grid.best_score_)
```

```python
model = grid.best_estimator_.fit(x, y)
feach_import = np.vstack((features, np.round(model.feature_importances_,2))).T
feach_import
```

Прогноз

```python
x_test = df_test[features].values
y_test = model.predict(x_test)

id_test = df_test.index.values

output = np.vstack([id_test, y_test]).T

output = pd.DataFrame(output)
output.columns = ['Id', 'Credit Default']
#output['Id'] = pd.to_numeric(output['Id'], downcast='integer')

output.to_csv('output.csv', index=False, sep= ',')
```
